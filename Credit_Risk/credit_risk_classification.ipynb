```python
# Import the modules
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, classification_report
```

---

## Split the Data into Training and Testing Sets

### Step 1: Read the `lending_data.csv` data from the `Resources` folder into a Pandas DataFrame.

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
file_path = Path("Resources/lending_data.csv")
df = pd.read_csv(file_path)

# Review the DataFrame
df
# df_loans.info()
```

### Step 2: Create the labels set (`y`)  from the “loan_status” column, and then create the features (`X`) DataFrame from the remaining columns.

```python
# Separate the data into labels and features

# Separate the y variable, the labels
y = df["loan_status"].values.reshape(-1,1)

# Separate the X variable, the features
X = df.drop(columns="loan_status")
```

```python
# Review the y variable Series
y
```

```python
# Review the X variable DataFrame
X
```

### Step 3: Split the data into training and testing datasets by using `train_test_split`.

```python
# Import the train_test_learn module
from sklearn.model_selection import train_test_split

# Split the data using train_test_split
# Assign a random_state of 1 to the function
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

---

## Create a Logistic Regression Model with the Original Data

###  Step 1: Fit a logistic regression model by using the training data (`X_train` and `y_train`).

```python
# Import the LogisticRegression module from SKLearn
from sklearn.linear_model import LogisticRegression

# Instantiate the Logistic Regression model
# Assign a random_state parameter of 1 to the model
model = LogisticRegression(solver='lbfgs', random_state=1)

# Fit the model using training data
model.fit(X_train, y_train)
```

### Step 2: Save the predictions on the testing data labels by using the testing feature data (`X_test`) and the fitted model.

```python
# Make a prediction using the testing data
predictions = model.predict(X_test)

prediction_list = predictions.tolist()

y_test_list = y_test.ravel().tolist()

# Out of curiousity: Compare predictions with actual values using a for loop
df_test = pd.DataFrame({
    "Predictions": prediction_list,
    "Actual Values": y_test_list
})

match_number = 0
non_match_number = 0
for i in df_test.index:
    if df_test["Predictions"][i] == df_test["Actual Values"][i]:
        match_number += 1
    else:
        non_match_number += 1
print(f"Number of matching predictions: {match_number}")
print(f"Number of non-matching predictions: {non_match_number}")

# Call the accuracy score, which should give the same result as the above analysis
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy score = {accuracy:.3f}")
```

### Step 3: Evaluate the model’s performance by doing the following:

* Generate a confusion matrix.

* Print the classification report.

```python
# Generate a confusion matrix for the model
confusion_matrix(y_test, predictions)
```

```python
# Paint a colorful confusion matrix
from IPython import display
display.Image("Resources/confusion_matrix.png", width = 500)
```

```python
# Print the classification report for the model
target_names = ["Healthy Loan", "High Risk of Default"]
print(classification_report(y_test, predictions, target_names=target_names, digits = 3))
```

### Step 4: Answer the following question.

**Question:** How well does the logistic regression model predict both the `0` (healthy loan) and `1` (high-risk loan) labels?

**Answer:** Of the health loans, virtually 100% were correctly predicted. The precision of the first direction - healthy loans - is 0.998, while the recall is 0.994. Thus, detection of healthy loans was very specific (precision was high) and accuracy was high as well (recall was also high).

However, the model did less well for the detection of high-risk loans. While the recall of high-risk loans was fairly high - only 34 out of 619 high-risk loans were missed by the model (i.e., became false negatives in this case) - the model struggled with precision; more than 15% of its predicted high-risk loans were actually health ones (i.e., false positives in this case).

In summary, the model does a better job at predicting healthy loans than high-risk loans. Additionally, if the goal of the model is to prioritize detecting high-risk loans (high recall), then it does well, at the expense of lower precision, i.e., unfairly classifying some health loans as false positives.
